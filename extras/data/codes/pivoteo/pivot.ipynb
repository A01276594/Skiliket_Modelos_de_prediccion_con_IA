```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from functools import reduce
```

```python
Device_id = 2
df = pd.read_csv(f"..\\..\\data\\devices\\device_{Device_id}\\Skiliket_Device_{Device_id}_FULL_HISTORY.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Time"] = pd.to_datetime(df["created_at"], errors="coerce")
df = df.dropna(subset=["created_at"]).set_index("Time")
```

```python
df_pivot = (
    df.pivot_table(
        index="Time",   # eje temporal
        columns="type",        # AQI, ECO2, etc.
        values="value",        # lo que quieres medir
        aggfunc="mean"         # por si hay duplicados
    )
    .sort_index()
)
```

```python
df_pivot.columns.name = None
```
